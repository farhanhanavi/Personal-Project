```python
most_picked_maps = pd.DataFrame(columns = ['event_id','map_picked'])
```

```python
map_picked = []
tournament   = []
for i, v in picks_df_astralis.iterrows():
    if (picks_df_astralis.loc[i,'team_1']  == 'Astralis') & (picks_df_astralis.loc[i,'t1_picked_1']  != '0.0'):
        x   = picks_df_astralis.loc[i,'t1_picked_1']
        x_i = picks_df_astralis.loc[i,'event_id']
        map_picked.append(x)
        tournament.append(x_i)
    elif (picks_df_astralis.loc[i,'team_2']  == 'Astralis') & (picks_df_astralis.loc[i,'t2_picked_1']  != '0.0'):
        y = picks_df_astralis.loc[i,'t2_picked_1']
        y_i = picks_df_astralis.loc[i,'event_id']
        map_picked.append(y)
        tournament.append(y_i)
    else:
        z = picks_df_astralis.loc[i,'left_over']
        z_i = picks_df_astralis.loc[i,'event_id']
        tournament.append(z_i)
        map_picked.append(z)

most_picked_maps = pd.DataFrame(tournament)  
most_picked_maps['map_picked'] = map_picked
most_picked_maps.groupby(0)['map_picked'].value_counts()
```
